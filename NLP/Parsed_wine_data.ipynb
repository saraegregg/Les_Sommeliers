```python
import pandas as pd


import os
import re
import nltk
import numpy as np


wine_df = pd.read_csv('https://raw.githubusercontent.com/saraegregg/Mod20_Group_Challenge/main/data_cleaning/cleaned_wine_data.csv', encoding= 'unicode_escape')
wine_df.head(10)
```

```python
wine_df.description = wine_df.description.apply(lambda x : x.replace(',',''))
wine_df.description = wine_df.description.apply(lambda x : x.replace('.',''))
wine_df.description = wine_df.description.apply(lambda x : x.lower())
wine_df.head(10)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
```

```python
token = wine_df["description"].apply(word_tokenize)
token
```

```python
wine_df["token"] = token
wine_df.head()
```

```python
lists = token
merged_list = []
for l in lists:
    merged_list += l
""
```

```python
stopWords = set(stopwords.words('english'))

wordsFiltered = []

for w in merged_list:
    if w not in stopWords:
        wordsFiltered.append(w)
""
```

```python
from nltk import FreqDist
fdist = FreqDist(wordsFiltered)
fdist.most_common(50)
```

```python
wine_df.rich = wine_df.description.apply(lambda x : 1 if x.find('ripe')>-1 else 0)
wine_df.rich.head(10)
```

```python
keywords = ['ripe', 'finish', 'aromas']
for k in keywords:
    wine_df[k] = wine_df.description.apply(lambda x : 1 if x.find(k)>-1 else 0)
wine_df.head()
```
